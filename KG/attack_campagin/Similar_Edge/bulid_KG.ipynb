```python
import pandas as pd
import networkx as nx
import json
from networkx.readwrite import json_graph

 
df = pd.read_csv('')   

G = nx.Graph()
 
for index, row in df.iterrows():
    node_id = int(row['Unnamed: 0'])   
    
     
    G.add_node(node_id, 
               name_version=str(row['name_version']),
               SourceCode_Path=str(row['SourceCode_Path']),
               Source=str(row['Source']),
               code_api=str(row['code_api']),
               n_tokens=int(row['n_tokens']),   
               embedding=str(row['embedding']),   
               cluster=str(row['Cluster'])
               )
    
 
for i in range(len(df)):
    for j in range(i + 1, len(df)):
        if df.at[i, 'Cluster'] == df.at[j, 'Cluster']:
            G.add_edge(int(df.at[i, 'Unnamed: 0']), int(df.at[j, 'Unnamed: 0']))

 
graph_data = json_graph.node_link_data(G)

 
with open('ruby_graph.json', 'w') as f:
    json.dump(graph_data, f, indent=4)
```

```python
import networkx as nx
import json
import os
from networkx.readwrite import json_graph

 
def save_subgraph(subgraph, folder_path, subgraph_id):
     
    subgraph_data = json_graph.node_link_data(subgraph)
    
     
    subgraph_filename = os.path.join(folder_path, f'subgraph_{subgraph_id}.json')
    
     
    with open(subgraph_filename, 'w') as f:
        json.dump(subgraph_data, f, indent=4)
    
    print(f"Subgraph {subgraph_id} saved to {subgraph_filename}")

 
with open('ruby_graph.json', 'r') as f:   
    graph_data = json.load(f)

 
G = json_graph.node_link_graph(graph_data)

 
connected_subgraphs = list(nx.connected_components(G))

 
output_folder = ''   

 
os.makedirs(output_folder, exist_ok=True)

 
for i, component in enumerate(connected_subgraphs):
    subgraph = G.subgraph(component).copy()   
    save_subgraph(subgraph, output_folder, i)

print("All connected subgraphs have been saved.")
```
